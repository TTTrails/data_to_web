```python
def max_and_min(data,max_year):
    max_all = 1
    min_all = 10000000
    for i in range(2002,max_year):
        index_max = max(data[i].values.tolist())
        index_min = min(data[i].values.tolist())
        
        if index_max > max_all:
            max_all = index_max
        if index_min < min_all :
            min_all = index_min
    return max_all,min_all
```
